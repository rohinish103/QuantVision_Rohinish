## Numpy Indexing and Selection

Now you will be given a few matrices, and be asked to replicate the resulting matrix outputs:

```python
arr1=np.array([list(range(1,26))]).reshape(5,5)
print(arr1)
```

```python
arr2=np.array([list(range(12,27))]).reshape(3,5)[:, :4]
print(arr2)
```

```python
print(arr1[3,4])
```

```python
print(arr1[:3,1:2])
```

```python
print(arr2[2,:])
```

```python
print(arr1[3:5 , :])
```
